```python
for column in data.columns:
    if column != 'ID':
        data[column].fillna(data[column].mean(), inplace=True)
```

# Reason why I used mean values of each column
#### There are various techniques employed to deal with null values within a dataset. Although deleting rows is a widely used method, it can result in data loss and inaccurate outcomes. A more effective approach is to substitute null values with the average values of the respective column. While this method is an estimation that introduces variability to the dataset, it helps prevent data loss and enhances the accuracy of data analysis

## Pie Chart showing the wine quality distribution

```python
quality_counts = data['quality'].value_counts()
exp = [0.2,0.2,0.2,0.3,0.8,0.6,1]
plt.pie(quality_counts, labels=quality_counts.index, startangle=35,explode =exp,shadow = True)
plt.axis('equal')
plt.title('Wine Quality Distribution')
plt.legend(title = 'Different values for distribution')
plt.show()
plt.savefig('quality_distribution_piechart.png')
```

```python
for column in data.columns:
    if column != 'ID' and column != 'quality':
        sb.barplot(x='quality', y= column, data = data,palette = 'rocket')
        plt.title(f'Quality graph for {column}')
        plt.figure(figsize=(12,10))

        plt.show()
```

```python
fig, axes = plt.subplots(4, 3, figsize=(20, 15))
for ax, column in zip(axes.flat, data.columns):
    if column != 'ID' and column != 'quality':
        sb.barplot(x='quality', y=column, data=data, palette='rocket', ax=ax)
        ax.set_title(f'Quality graph for {column}')
        ax.set_xlabel('Quality')
        ax.set_ylabel(column)


plt.tight_layout()
plt.show()
plt.savefig('quality_check.png')
```
